```python
import cv2
# Load a pretrained YOLOv8n model
model = YOLO('runs/classify/train/weights/best.pt')
# Gets the image at the specified path as a numpy array
def getImage(path):
    img = cv2.imread(path)
    return img

# Displays the results of model(image)
def getResults(results):
    for r in results:
        guesses = r.probs.data
        if guesses[0] == max(guesses):
            return "bolt"
        if guesses[8] == max(guesses):
            return "tag"
        aggregate = (8 * guesses[1] + 5 * guesses[2] + 4 * guesses[3] + 9 * guesses[4] + 
                     1 * guesses[5] + 7 * guesses[6] + 6 * guesses[7] + 10 * guesses[9] + 3 * guesses[10] + 2 * guesses[11]) 
        return float(aggregate)
```

```python
im = getImage("easytest.jpg")
results = model.predict(im)
print(getResults(results))
im = getImage("medtest.jpg")
results = model.predict(im)
print(getResults(results))
im = getImage("hardtest.jpg")
results = model.predict(im)
print(getResults(results))
im = getImage("tagtest.jpg")
results = model.predict(im)
print(getResults(results))
```
